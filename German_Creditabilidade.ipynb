```python
import pandas as pd

dados = pd.read_csv('german_Credit_Data.csv', sep=';')
dados.head()
```

```python
mapa = {'Creditability': 'credibilidade',
        'Account_Balance': 'saldo_na_conta',
        'Duration_of_Credit_monthly': 'duracao_do_credito_mensal',
        'Payment_Status_of_Previous_Credit': 'Estado_de_pagamento_do_crédito_anterior',
        'Purpose': 'Proposito',
        'Credit_Amount': 'Limite_de_credito',
        'Value_Savings_Stocks': 'Acoes_de_Economia_de_valor',
        'Length_of_current_employment': 'Duracao_do_emprego_atual',
        'Instalment_per_cent': 'Porcentagem_de_Parcela',
        'Sex_Marital_Status': 'Estado_civil',
        'Duration_in_Current_address': 'Duracao_no_endereco_atual',
        'Most_valuable_available_asset': 'Ativo_mais_valioso_disponivel',
        'Age_years': 'Idade',
        'Concurrent_Credits': 'Credito_Simultaneos',
        'Type_of_apartment': 'Tipo_do_apartamento',
        'No_of_Credits_at_this_Bank': 'Numeros_de_credito_nesse_banco',
        'Occupation': 'Ocupacao',
        'No_of_dependents': 'Numeros_de_dependentes',
        'Telephone': 'Telefone',
        'Foreign_Worker': 'Trabalhador_estrangeiro'}

dados = dados.rename(columns = mapa)
dados.head()
```

```python
# Classificação com o Linear SVC

y = dados['credibilidade'] # classe: 1 para sim, pagou o emprestimo; 0 para não, não pagou o emprestimo 
x = dados[['Limite_de_credito','Idade']]

from sklearn.model_selection import train_test_split # Separar em treino e teste
from sklearn.svm import LinearSVC # pra traçar uma reta linear
from sklearn.metrics import accuracy_score # fornecer as metricas 
import numpy as np

SEED = 5
np.random.seed(SEED)
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.25,
                                                        stratify = y)

print(f'treinaremos com {len(treino_x)} elementos e testaresmo com {len(teste_x)} elementos')

# Instaciando o modelo
modelo = LinearSVC()

# Treinando o modelo
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print(f"A acuracia foi de: {acuracia:.2f}%")
```
