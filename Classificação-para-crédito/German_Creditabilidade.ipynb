```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt

dados = pd.read_csv('german_Credit_Data.csv', sep=';')
dados.head()
```

```python
mapa = {'Creditability': 'credibilidade',
        'Account_Balance': 'saldo_na_conta',
        'Duration_of_Credit_monthly': 'duracao_do_credito_mensal',
        'Payment_Status_of_Previous_Credit': 'Estado_de_pagamento_do_crédito_anterior',
        'Purpose': 'Proposito',
        'Credit_Amount': 'Limite_de_credito',
        'Value_Savings_Stocks': 'Acoes_de_Economia_de_valor',
        'Length_of_current_employment': 'Duracao_do_emprego_atual',
        'Instalment_per_cent': 'Porcentagem_de_Parcela',
        'Sex_Marital_Status': 'Estado_civil',
        'Duration_in_Current_address': 'Duracao_no_endereco_atual',
        'Most_valuable_available_asset': 'Ativo_mais_valioso_disponivel',
        'Age_years': 'Idade',
        'Concurrent_Credits': 'Credito_Simultaneos',
        'Type_of_apartment': 'Tipo_do_apartamento',
        'No_of_Credits_at_this_Bank': 'Numeros_de_credito_nesse_banco',
        'Occupation': 'Ocupacao',
        'No_of_dependents': 'Numeros_de_dependentes',
        'Telephone': 'Telefone',
        'Foreign_Worker': 'Trabalhador_estrangeiro'}

dados = dados.rename(columns = mapa)
dados.head()
```

```python
# Verificar se os dados que quero classificar estão balanceados
import seaborn as sns
%matplotlib inline
ax = sns.countplot(data = dados, x = 'credibilidade')
```

```python
from imblearn.over_sampling import SMOTE
```

Pode-se perceber que os dados não estão balanceados: É interessante utilizar uma estratégia de Oversampling

Para isso podemos aplicar o SMOTE, devemos separar  os dados em variáveis características (labels) e resposta (classe) 

Smote cria observações intermediárias entre os dados que estão próximos, ou seja, ele cria observações pra classe minoritária que para esse caso é (0: pessoas que não pagaram o emprestimo)


```python
x = dados.drop(['credibilidade'], axis = 1)
y = dados['credibilidade']

smt = SMOTE(random_state=123)  # Instancia um objeto da classe SMOTE
x, y = smt.fit_resample(x, y)  # Realiza a reamostragem do conjunto de dados
dados = pd.concat([y, x], axis = 1)
dados.head()
```

```python
ax = sns.countplot(x='credibilidade', data = dados)  # plotando a variável target balanceada.
```

```python
# Mostra uma matriz de correlação entre todas as colunas númericas
dados.corr()
# Plota usando cores para distinguir forte correlação proporcional ou inversamente proporcional
sns.heatmap(dados.corr())
```

```python
# Normalização das labels
scaler = MinMaxScaler().fit(dados)
dados_norm = scaler.fit_transform(dados)
dados_norm = pd.DataFrame(dados_norm)
dados_norm.columns = dados.columns
dados_norm.head()
```

```python
dados_norm.shape
```

```python
import seaborn as sns
y = dados_norm['credibilidade']
X = dados_norm.drop(['credibilidade'], axis = 1)
X.head()
```

```python
# Métodos pra seleçao das melhores features
from sklearn.feature_selection import SelectKBest, chi2

# O Dataset possui 20 features / colunas

# Aplica a seleção de características usando o teste chi-quadrado
kbest = SelectKBest(chi2, k=8)
X_new = kbest.fit_transform(X, y)

# Obtém as posições das colunas selecionadas
posicoes_colunas = kbest.get_support()

# Obtém os nomes das colunas originais correspondentes
nomes_colunas = X.columns[posicoes_colunas]

nomes_colunas
```

```python
# sem normalizar estava sendo 'duracao_do_credito_mensal' e 'Limite_de_credito'
sns.scatterplot(data = dados_norm, y = 'duracao_do_credito_mensal', x = 'Limite_de_credito', hue = y)
```

```python
sns.relplot(data = dados, y = 'duracao_do_credito_mensal', x = 'Limite_de_credito', hue = y, col = y )
```

```python
# Classificação com o Linear SVC
#limite de credito mensal e limite de credito
y = dados_norm['credibilidade'] # classe: 1 para sim, pagou o emprestimo; 0 para não, não pagou o emprestimo 
x = dados_norm[['saldo_na_conta','Estado_de_pagamento_do_crédito_anterior']]

from sklearn.model_selection import train_test_split # Separar em treino e teste
from sklearn.svm import LinearSVC # pra traçar uma reta linear
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.metrics import accuracy_score # fornecer as metricas 
import numpy as np

treino_x, teste_x, treino_y, teste_y = train_test_split(X, y, test_size = 0.25, random_state = 123,
                                                        stratify = y)

print(f'treinaremos com {len(treino_x)} elementos e testaresmo com {len(teste_x)} elementos')

# Instaciando o modelo
modelo = LinearSVC()

# Treinando o modelo
modelo.fit(treino_x, treino_y)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes) * 100
print(f"A acuracia foi de: {acuracia:.2f}%")
```
